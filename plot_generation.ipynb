```python
import numpy as np
import matplotlib.pyplot as plt

# Exponential plot
x = np.linspace(0, 10, 100)
y = np.exp(x)
plt.plot(x, y)
plt.title("Exponential Function")
plt.savefig("assets/exp-plot.png")
plt.clf()

# Quadratic plot
y = x**2
plt.plot(x, y)
plt.title("Quadratic Function")
plt.savefig("assets/quad-plot.png")
```
